```python
import matplotlib.pyplot as mp
import numpy as np


print("Hey! What's on your mind?")
com1 = input()

stud = {}
sub = ["Hindi","English","Maths","Social","Science"]
mar = []
top="" #topper name
sct=0 #student count
tc=0 #total count
htp="" #hindi top
etp="" #english top
sctp="" #science top
sotp="" #social top
mtp="" #math top
ctot=0 #class total
cavg=0 #class average(All subjects)
Pass=0 #no. of students who have passed their exams
fail=0 #no. of students who have failed in exams
faild={} #dictionary of failed students
fs=[] #
hf=0
ef=0
sof=0
scf=0
mf=0

while com1.lower()!="exit":





    

    
    #ADDING A STUDENT
    
    if com1.lower()=="add a student" or com1.lower()=="add student" or com1.lower()=="add":
        stdn=input("Enter the name of student")
        if stdn in stud:
            print("Student already exists in the system")
            

        else:
            stud[stdn]={}
            stud[stdn]["tot"]=0
            stud[stdn]["per"]=0
            stud[stdn]["grd"]=""
            tot=0
            faild[stdn]=[]
            for i in sub:
                mar1 = int(input("Please enter marks in " + i))
                tot+=mar1
                stud[stdn][i]=mar1
                
                if mar1<33:
                    stud[stdn]["grd"]="Fail"
                    faild[stdn].append(i)
                    
                

                        

                elif mar1>=33 and stud[stdn]["grd"]!="Fail":
                    stud[stdn]["grd"]="Pass"


                if i=="Hindi":
                    hf+=mar1

                elif i=="English":
                    ef+=mar1

                elif i=="Social":
                    sof+=mar1


                elif i=="Science":
                    scf+=mar1

                elif i=="Maths":
                    mf+=mar1




            if stud[stdn]["grd"]=="Pass":
                Pass+=1

            else:
                fail+=1
                fs.append(stdn)

            sct+=1
            ctot+=tot

                    
            stud[stdn]["tot"]=tot
            stud[stdn]["per"]=tot/5

            cavg=(ctot)/sct
            

            
            if len(stud)==1:
                htp=stdn
                etp=stdn
                sctp=stdn
                sotp=stdn
                mtp=stdn
                

            else:
                if stud[stdn]["Hindi"]>stud[htp]["Hindi"]:
                    htp=stdn


                if stud[stdn]["English"]>stud[etp]["English"]:
                    etp=stdn

                if stud[stdn]["Science"]>stud[sctp]["Science"]:
                    sctp=stdn

                if stud[stdn]["Social"]>stud[sotp]["Social"]:
                    sotp=stdn

                if stud[stdn]["Maths"]>stud[mtp]["Maths"]:
                    mtp=stdn


            com1 = input("Enter a command")

                



    

    
    #COMPARING MARKS OF TWO STUDENTS

    elif com1.lower()=="compare students" or com1.lower()=="compare":
        if len(stud)>1:
            n1=input("Enter the name of student 1")
            n2=input("Enter the name of student 2")
            if n1 in stud and n2 in stud:
                for i in sub:
                    print(i,stud[n1][i],stud[n2][i])
                    l1=[stud[n1][i] for i in sub]
                    l2=[stud[n2][i] for i in sub]
                    w=0.4
                    x1=np.arange(len(l1))
                    x2=[i+w for i in x1]

                mp.bar(x1,l1,w,label=n1)
                mp.bar(x2,l2,w,label=n2)
                mp.xticks(x1+w/2,sub)
                mp.show()
                

            else:
                print("Either of the two students is not in our data")

            

        else:
            print("There is only one student in our data, please add one or more student to compare")


        com1 = input("Enter a command")
            





    




    #REPORT OF A STUDENT

    elif com1.lower()=="report":
        if len(stud)>0:
            rn = input("Please enter the name of student")
            if rn in stud:
                for i in sub:
                    print(i,stud[rn][i])

                print("Total =",stud[rn]["tot"])
                print("Percentage =",stud[rn]["per"])
                print(stud[rn]["grd"])
                rl=[]
                for i in sub:
                    rl.append(stud[rn][i])

                mp.bar(sub,rl)
                mp.show()
                

            else:
                print("Student not found")
            

        else:
            print("No student found in our data")


        

        com1=input("Enter a command")
        






    


    
    #FINDING THE TOPPER

    elif com1.lower()=="topper":
        if len(stud)>0:
            tp=0
            for i in stud:
                
                if stud[i]["tot"]>tp:
                    tp=stud[i]["tot"]
                    top=i
        
            print(top)
            

        else:
            print("No student found in our data")

        com1=input("Enter a command")






    



    #FINDING THE SUBJECT TOPPER

    elif com1.lower()=="stopper":
        if len(stud)>0:
            sbt = input("Please enter the subject whose topper you would want to find")
            if sbt.lower()=="hindi":
                print(htp)

            elif sbt.lower()=="english":
                print(etp)


            elif sbt.lower()=="social":
                print(sotp)

            elif sbt.lower()=="science":
                print(sctp)

            elif sbt.lower()=="maths":
                print(mtp)

            else:
                print("Please type a valid subject name")


        else:
            print("No student found in our data")


        com1 = input("Enter a command")




    



    

    #CLASS AVG

    elif com1.lower()=="avg":
        print("Class average is",round(ctot/sct,2),"marks")
        print("Average marks in Hindi is",round(hf/sct,2))
        print("Average marks in English is",round(ef/sct,2))
        print("Average marks in Social is",round(sof/sct,2))
        print("Average marks in Science is",round(scf/sct,2))
        print("Average marks in Maths is",round(mf/sct,2))
        com1=input("Enter a command")






    

    

    #CLASS PASS-FAIL PIE CHART

    elif com1.lower()=="performance chart" or com1.lower()=="pchart":
        x=[Pass,fail]
        y=["Pass","Fail"]

        mp.pie(x,labels=y)
        mp.show()
        com1=input("Enter a command")







    


    #DELETE A STUDENT
    elif com1.lower()=="delete":
        if len(stud)>0:
            dn = input("Please enter the name of student you would want to delete")
            cn = input("Are you sure you?")
            if cn.lower()=="yes" and dn in stud:
                if stud[dn]["grd"]=="Pass":
                    Pass-=1 

                else:
                    fs.remove(dn)
                    del faild[dn]
                    fail-=1

                sct-=1
                hf-=stud[dn]["Hindi"]
                ef-=stud[dn]["English"]
                scf-=stud[dn]["Science"]
                sof-=stud[dn]["Social"]
                mf-=stud[dn]["Maths"]

                
                


                    
                del stud[dn]
                print("Student deleted successfully")



                if htp==dn:
                    htp=""
                    htpm=0
                    for i in stud:
                        if stud[i]["Hindi"]>htpm:
                            htpm=stud[i]["Hindi"]
                            htp=i


                if etp==dn:
                    etp=""
                    etpm=0
                    for i in stud:
                        if stud[i]["English"]>etpm:
                            htpm=stud[i]["English"]
                            etp=i

                if sotp==dn:
                    sotp=""
                    sotpm=0
                    for i in stud:
                        if stud[i]["Social"]>sotpm:
                            sotpm=stud[i]["Social"]
                            sotp=i

                if sctp==dn:
                    sctp=""
                    sctpm=0
                    for i in stud:
                        if stud[i]["Science"]>sctpm:
                            sctpm=stud[i]["Science"]
                            sctp=i

                if mtp==dn:
                    mtp=""
                    mtpm=0
                    for i in stud:
                        if stud[i]["Maths"]>mtpm:
                            mtpm=stud[i]["Maths"]
                            mtp=i


            elif dn not in stud:
                print("Student not found in our data")


            com1=input("Enter a command")





    




    #UPDATE MARKS
    elif com1.lower()=="update":
        if len(stud)>0:
            un = input("Enter name of the student")
            sb = input("Enter name of the subject")
            nm = int(input("Enter the score of the student"))
            st=stud[un][sb]
            if un in stud and sb in stud[un]:
                t=stud[un][sb]
                stud[un]["tot"]-=stud[un][sb] #subtracting old marks to update new total
                ctot-=stud[un][sb]  #subtracting old marks to update new class total
                stud[un][sb]=nm  #Assigned new score to the subject
                stud[un]["tot"]+=stud[un][sb] #updated student total
                stud[un]["per"] = stud[un]["tot"] / 5 #updated percentage
                ctot+=stud[un][sb] #class total updated
                if sb=="Hindi":
                    hf-=t
                    hf+=stud[un][sb]

                if sb=="English":
                    ef-=t
                    ef+=stud[un][sb]

                if sb=="Science":
                    scf-=t
                    scf+=stud[un][sb]

                if sb=="Social":
                    sof-=t
                    sof+=stud[un][sb]

                if sb=="Maths":
                    mf-=t
                    mf+=stud[un][sb]


                
                print("Score updated successfully")

                ogrd=stud[un]["grd"]
                
                stud[un]["grd"]=""
                for i in sub:
                    if stud[un][i]<33:
                        stud[un]["grd"]="Fail"
                        

                    elif stud[un][i]>=33 and stud[un]["grd"]!="Fail":
                        stud[un]["grd"]="Pass"
                        if un in faild:
                            del faild[un]


                if ogrd=="Fail" and stud[un]["grd"]=="Pass":
                    fs.remove(un)
                    fail-=1
                    Pass+=1


                elif ogrd=="Pass" and stud[un]["grd"]=="Fail":
                    fs.append(un)
                    Pass-=1
                    fail+=1
                

            
            else:
                print("Incorrect student name or subject")


        else:
            print("No student found in our data")

        com1=input("Enter a command")












    
    #NO. OF FAILED STUDENTS AND THE SUBJECTS THEY'VE FAILED IN
    
    elif com1.lower()=="analysis":
        if len(fs)>0:
            
            print("Below are the students who have failed in the exams")
            for i in fs:
                print(i)

            sa = input("Please enter the name of students you would want to check for(Enter NONE if you would like to skip this step)")
            if sa.lower()!="none":
                for i in faild[sa]:
                    print("")
                    print("")
                    print("Below is the list of subjects the student has failed in")
                    print(i)

        else:
            print("There is no student who has failed in the exam")

        com1=input("Enter a command")








    #HELP

    elif com1.lower()=="help":
        print("The commands listed below can be used in the software.")
        print("")
        print("")
        print("ADD  |  COMPARE  |  DELETE  |  UPDATE  |  ANALYSIS  |  PCHART  |  STOPPER  |  TOPPER  |  REPORT  |  AVG")
        com1=input("Enter a command")
    
    else:
        print("Please opt for a valid option")
        print("Type 'Help' if you need help with the commands")
        com1 = input("Enter a command")
        
```
